````python
#  Import and read the csv.
import pandas as pd 

# Set DataFrame
nyc_df = pd.read_csv(io.BytesIO(uploaded['dropped_na_nyc_color_fix.csv']))
nyc_df
````

````python
# Replace all 1900's values to 1900 together
def change_year(year):
  if year <= 1999: 
    return 1990
  elif year >= 2000 and year <= 2009 :
    return 2000
  elif year >= 2010 and year <= 2018 :
    return 2010
  elif year > 2018:
    return 2100
  else: return year;

nyc_df['Vehicle_Year']= nyc_df['Vehicle_Year'].apply(change_year)

Vehicle_Year_count=nyc_df['Vehicle_Year'].value_counts()
Vehicle_Year_count
  
````

````python
# drop Vehicle_Year rows with 2100's
nyc_df= nyc_df[nyc_df.Vehicle_Year != 2100]  
nyc_df
````

````python
# drop commercial vehicles
nyc_df= nyc_df[nyc_df.Plate_Type != 'COM']  
nyc_df
````

````python
# cut down tickets 
def cut_tickets(ticket):
  if ticket == 21.0:
    return 'Street Cleaning'
  elif ticket == 36.0:
    return 'Speeding'
  else: return 'other'

nyc_df['Violation_Code']= nyc_df['Violation_Code'].apply(cut_tickets)
nyc_df
````

````python
# Look at 'Vehicle_Color' value counts for binning
Vehicle_Color_count=nyc_df['Vehicle_Color'].value_counts()
Vehicle_Color_count
````

````python
# Group colors together:
def group_colors(color):
  if color == 'GREY':
    return "DULL"
  elif color == 'WHITE':
    return "DULL"
  elif color == 'BLACK':
    return "DULL"
  elif color == 'SILVER':
    return "DULL"
  elif color == 'BLUE':
    return "BRIGHT"
  elif color == 'RED':
    return "BRIGHT"
  elif color == 'GREEN':
    return "BRIGHT"
  elif color == 'YELLOW':
    return "BRIGHT"
  elif color == 'ORANGE':
    return "BRIGHT"
  elif color == 'PURPLE':
    return "BRIGHT"

nyc_df['Vehicle_Color']= nyc_df['Vehicle_Color'].apply(group_colors)


# Look at 'Vehicle_Color' value counts for binning
Vehicle_Color_count=nyc_df['Vehicle_Color'].value_counts()
Vehicle_Color_count
````

````python
# Determine the number of unique values in each column.
nyc_df.nunique()
````

````python
# Generate our categorical variable lists
cat_objects = nyc_df.dtypes[nyc_df.dtypes == "object"].index.tolist()
cat_objects

````

````python
# Create a OneHotEncoder instance
enc = OneHotEncoder(sparse=False)

# Fit and transform the OneHotEncoder using the categorical variable list
encode_df = pd.DataFrame(enc.fit_transform(nyc_df[cat_objects]))

# Add the encoded variable names to the dataframe
encode_df.columns = enc.get_feature_names(cat_objects)
encode_df.head()
````

````python
# Merge one-hot encoded features and drop the originals
nyc_df = nyc_df.merge(encode_df,left_index=True, right_index=True)
nyc_df = nyc_df.drop(cat_objects,1)
nyc_df.head()


````

````python
# Split our preprocessed data into our features and target arrays
y = nyc_df['Violation_Code_Street Cleaning'].values
X = nyc_df.drop(["Violation_Code_Street Cleaning", 'Violation_Code_Speeding'],1).values

# Split the preprocessed data into a training and testing dataset
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=78)
````

````python
# Create a StandardScaler instances
scaler = StandardScaler()

# Fit the StandardScaler
X_scaler = scaler.fit(X_train)

# Scale the data
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
````

### Deliverable 2: Compile, Train and Evaluate the Model

````python
# Define the model - deep neural net, i.e., the number of input features and hidden nodes for each layer.
number_input_features = len(X_train[0])
hidden_nodes_layer1 = 10
hidden_nodes_layer2 = 8
hidden_nodes_layer3 = 5


nn = tf.keras.models.Sequential()

# First hidden layer
nn.add(
    tf.keras.layers.Dense(units=hidden_nodes_layer1, input_dim=number_input_features, activation="leaky_relu")
)

# Second hidden layer
nn.add(tf.keras.layers.Dense(units=hidden_nodes_layer2, activation="leaky_relu"))

# Third hidden layer
nn.add(tf.keras.layers.Dense(units=hidden_nodes_layer3, activation="sigmoid"))


# Output layer
nn.add(tf.keras.layers.Dense(units=1, activation="sigmoid"))

# Check the structure of the model
nn.summary()
````

````python
# Import checkpoint dependencies
import os
from tensorflow.keras.callbacks import ModelCheckpoint

# Define the checkpoint path and filenames
os.makedirs("checkpoints/",exist_ok=True)
checkpoint_path = "checkpoints/weights.{epoch:02d}.hdf5"
````

````python
# Compile the model
nn.compile(loss="binary_crossentropy", optimizer="adam", metrics=["accuracy"])

# Create a callback that saves the model's weights every 5 epoch
cp_callback = ModelCheckpoint(
    filepath=checkpoint_path,
    verbose=1,
    save_weights_only=True,
    save_freq=5)
````

````python
# Train the model
fit_model = nn.fit(X_train_scaled,y_train,epochs=30,callbacks=[cp_callback])
````

````python
# Evaluate the model using the test data
model_loss, model_accuracy = nn.evaluate(X_test_scaled,y_test,verbose=2)
print(f"Loss: {model_loss}, Accuracy: {model_accuracy}")
````
